## Goal <a class="anchor" id="goal"></a>

The goal of this chapter is to prove two important theorems of finance by generalizing the binomial models seen in the previous lessons.

$\underline{TL;DR:}$
1. Given a fixed numeraire, the market is free of arbitrage if and only if there exists a martingale measure.
2. Assume that the market is free of arbitrage, then the market is complete if and only if the martingale measure is unique.
3. Use this to price a derivative with respect to such a martingale.

## Model setup <a class="anchor" id="model_setup"></a>

- Two points in time $t=0, 1$
- Probability space $(\Omega, \mathcal{F}, \mathbb{P})$, with
    - $\Omega = \{ \omega_{1}, ..., \omega_{M} \}$
    - $\mathbb{P}$ defined on $2^{\Omega}$
- $N+1$ assets and we will denote the price of asset number $i$ at time t by: $$ S_{t} = (S_{t}^{0}, ..., S_{t}^{N}) $$
    - $S_{0}$ being deterministic
    - $S_{1}\in \mathcal{F}$
    - Strictly positive asset $S^{0}$ as a numeraire asset.

# Arbitrage <a class="anchor" id="arbitrage"></a>

## Normalized process <a class="anchor" id="normalized_process"></a>

Using the asset 0 as a numeraire will allow us to compare money today, $t = 0$, to money in the future, $t = 1$.

<b>Definition:</b> The normalized price process is defined by: 

$$
Z_{t}=\frac{S_{t}}{S^{0}_{t}} = (\frac{S_{t}^{0}}{S_{t}^{0}}, ..., \frac{S_{t}^{N}}{S_{t}^{0}})
$$

<u>Remark:</u> If $S_{0}$ is a bank account we are simply discounting everything to present value.
***

## Martingale measure <a class="anchor" id="martingale_measure"></a>

<b>Proposition: </b>The market is free of arbitrage if and only if there exists a probability measure $Q$ on $\Omega$ such that 
- $Q(\omega)>0 \text{ for all } \omega \in \Omega$
- $\frac{S^{i}_{0}}{S^{0}_{0}} = E^{Q}[\frac{S^{i}_{1}}{S^{0}_{1}}] \text{ for all } i\geq 1$

Such a measure Q is called a martingale measure.

<u>Notes:</u>
- The second condition means that the normalized price process $Z_{t}$ is a martingale under Q. Note that different numeraire assets will give rise to different martingale measures.
- The larger N is the easier it is to find a martingale measure Q.

To do the proof we will use Farkas’ Lemma, which is a special case of the separation theorem for convex sets.

<b>Definition: (Equivalent probability measures)</b> Two probability measures P and Q on $(\Omega, \mathcal{F})$ are equivalent (P ~ Q) if

$$
P(A)=0 \Leftrightarrow Q(A)=0 \text{ for all } A \in \mathcal{F}
$$

<b>Definition: (Martingale measure)</b> Consider the market above and fix the asset $S^{0}$ as the numeraire asset. A probability measure $Q$ on $(\Omega, \mathcal{F})$ is said to be a martingale measure if:
- $Q \sim P$
- For every $i = 1, ..., N$ the normalized price process

$$
Z_{t}^{i}= \frac{S_{t}^{i}}{S^{t}_{0}}
$$

is a martingale under $Q$.

## First Fundamental Theorem <a class="anchor" id="first_theorem"></a>

<b>Theorem (First Fundamental Theorem)</b>: Given a fixed numeraire:

<center>Market is free of arbitrage $\Leftrightarrow$ there exists a Martingale measure.</center>
***

# Completeness <a class="anchor" id="completeness"></a>

## Portfolio <a class="anchor" id="portfolio"></a>

<b>Definition: (Portfolio)</b> Let's define a <i>portfolio</i> $h = (h^{0}, ..., h^{N})^{T}$, where the super-index T denotes transpose.

- <i>Value process</i> $V^{S}$ of the portfolio as: $$ V^{S}(t) = \sum_{i=0}^{N} h_{i}S^{i}_{t} $$ Therefore the normalized value process $V^{Z}$ is $$ V^{S}(t) = \sum_{i=0}^{N} h_{i}Z^{i}_{t} = \frac{V^{S}(t)}{S_{t}^{0}} $$

- An <i>arbitrage portfolio</i> is a portfolio $h$ such that
1. $V^{S}(0) = 0$
2. $V^{S}(1) \geq 0 \text{ with probability 1 }$
3. $P(V^{S}(1)>0)>0$

## Reachability <a class="anchor" id="reachability"></a>

We continue by studying completeness of the market, i.e. if all derivatives are reachable. The formal definitions are given below.

<b>Definition:</b> A financial derivative or cotingent claim is a any random variable defined on $(\Omega, \mathcal{F})$. A claim $X$ is said to be reachable if there exists a portfolio $h$ such that  

$$
\mathcal{P}(V_{1}^{S}=X) = 1
$$

The portfolio $h$ is said to be a hedging or replicating portfolio for the claim $X$.

<b>Proposition:</b> The market is complete $\Leftrightarrow$ $A$ has rank $M$.

<u>Remark:</u> The rows of $A$ is the price vector $S_{1}$ for the different outcomes $\omega_{i}$

## Second fundamental theorem <a class="anchor" id="second_theorem"></a>

<b>Theorem 2 (Second Fundamental Theorem):</b> Assume that the market is free of arbitrage,

<center>Market is complete $\Leftrightarrow$ the martingale measure is unique</center>

<u>Remark:</u> There are 0,1 or ∞ many martingale measures.

# Pricing <a class="anchor" id="pricing"></a>

## Context <a class="anchor" id="context"></a>

- Unnormalized economy
- We work with a fixed numeraire asset $S_{0}$
- Assumption: The model is free of arbitrage
- A claim X is reachable by the replicating portfolio h

Ok. So. The goal is to develop the price of the derivative at time t = 0 so that it is linked to the price at time t = 1 (specifically a martingale).

Compute the derivative price at t=0 using:
1. The replicating portfolio $$ \Pi(0; X) = V^{h}(0) = \sum_{i=0}^{N}h^{i}S_{0}^{i} $$
2. The martingale respect of the normalized value process $V^{Z}$: $$ \frac{S^{i}_{0}}{S^{0}_{0}} = E^{Q}[\frac{S^{i}_{1}}{S^{0}_{1}}] \forall i\geq 1 $$
We have: $$ \Pi(0;X) = V^{h}(0) = \sum_{i=0}^{N}h^{i}S_{0}^{i} = \sum_{i=0}^{N}h^{i} S_{0}^{0} E^{Q}[\frac{S_{1}^{i}}{S_{1}^{0}}]$$

3. Linearity of the expectancy + definition of the value process + h replicates X  $$
\Pi(0;X) = S_{0}^{0}E^{Q}[\sum_{i=0}^{N}h^{i}\frac{S_{1}^{i}}{S_{1}^{0}}] = S_{0}^{0}E^{Q}[\frac{V_{1}^{h}}{S_{1}^{0}}] = S_{0}^{0}E^{Q}[\frac{X}{S_{1}^{0}}] = S_{0}^{0}E^{Q}[\frac{\Pi(1;X)}{S_{1}^{0}}]$$



<b>CONCLUSION:</b> The normalized price of the claim is also a martingale under Q!

<b>Proposition:</b> Suppose that the market is free of arbitrage. Then the following hold:

1. To extend the market with X and keep it free of arbitrage we have to set $$ \Pi(0;X) = S_{0}^{0}E^{Q}[\frac{X}{S_{1}^{0}}] $$ for some martingale measure Q.

2. If X is not reachable different choices of martingale measures for the fixed numeraire give different prices.

3. If X is reachable (2) will not depend on the choice of martingale measure Q and $$ \Pi(0;X) = V^{h}(0) \text{ for all h replicating X} $$
